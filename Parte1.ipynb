PANDAS

```python
import pandas as pd
```

INCIO CON LA LIMPIEZA DE LA DATA 1

```python
votosobtenidos=pd.read_html(votosobtenidosLink,match='Distrito')
votosobtenidos[0] # cual
```

```python
votosobtenidos=votosobtenidos[0].copy() # always
```

```python
votosobtenidos.info()
```

Reviso las columnas para ver cuáles retirar

```python
votosobtenidos.columns[1::]
```

```python
#convertir columnas
for col in votosobtenidos.columns[1::]:
    votosobtenidos[col] = votosobtenidos[col].astype(str)
```

```python
votosobtenidos[votosobtenidos.columns[1::]] = votosobtenidos.loc[:, votosobtenidos.columns[1::]].apply(
    lambda x: x.str.replace('\s+', '', regex=True)
).apply(
    lambda x: pd.to_numeric(x, errors='coerce')
)
```

```python
votosobtenidos.info()
```

Elegir las variables deseadas (DISTRITO Y RP)

```python
votosobtenidos=votosobtenidos.loc[:,['Distrito','RP']][:-1]
votosobtenidos
```

LIMPIEZA DATA 2 (según data 1)

```python
percapitaLink='https://github.com/josesotoxi/practica/raw/refs/heads/main/data2_percapita.csv'
```

```python
import pandas as pd
percapita=pd.read_csv(percapitaLink)
```

Corroborar código para colocar cada palabra en mayúscula

```python
data = {'Distrito': ['ancón', 'ate', 'barranco']}
df = pd.DataFrame(data)
print("DataFrame original:")
print(df)

df['Distrito'] = df['Distrito'].str.capitalize()
print("\nDataFrame después de capitalize():")
print(df)

data = {'Distrito': ['ancón', 'ate', 'barranco']}
df = pd.DataFrame(data)

df['Distrito'] = df['Distrito'].str.title()
print("\nDataFrame después de title():")
print(df)
```

```python
percapita['Distrito'] = percapita['Distrito'].str.title()
```

```python
percapita['Distrito'] = percapita['Distrito'].str.title()
```

corregir casos específicos

```python
percapita['Distrito'] = percapita['Distrito'].replace({'San Juan De Miraflores':'San Juan de Miraflores'})
percapita['Distrito'] = percapita['Distrito'].replace({'San Juan De Lurigancho':'San Juan de Lurigancho'})
percapita['Distrito'] = percapita['Distrito'].replace({'Magdalena Del Mar':'Magdalena del Mar'})
percapita['Distrito'] = percapita['Distrito'].replace({'San Martín De Porres':'San Martín de Porres'})
percapita['Distrito'] = percapita['Distrito'].replace({'Santa María Del Mar':'Santa María del Mar'})
percapita['Distrito'] = percapita['Distrito'].replace({'Santiago De Surco':'Santiago de Surco'})
percapita['Distrito'] = percapita['Distrito'].replace({'Villa María Del Triunfo':'Villa María del Triunfo'})
```

```python
electoresLink='https://github.com/josesotoxi/practica/raw/refs/heads/main/votosfinal.csv'
import pandas as pd
electores=pd.read_csv(electoresLink)
```

COMBINAR DATA

```python
electorescapita = pd.merge(electores, percapita, on='Distrito')
```

```python
final.reset_index(drop=True,inplace=True)
final
```

Renombrar variable

```python
final.rename(columns={'Ingreso Per Capita':'percap'},inplace=True)
```

cambiar comas por puntos

```python
final['percap'] = final['percap'].astype(str).str.replace(',', '.', regex=False)
```

```python
final['percap'] = pd.to_numeric(final['percap'], errors='coerce')
```
